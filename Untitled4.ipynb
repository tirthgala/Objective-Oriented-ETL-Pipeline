```python
import pydot
from IPython.display import Image, display

# Define the graph
graph = pydot.Dot(graph_type='graph')

# Add nodes
nodes = {
    "Team": pydot.Node("Team", shape='ellipse', style='filled', fillcolor='lightblue'),
    "Match": pydot.Node("Match", shape='ellipse', style='filled', fillcolor='lightblue'),
    "Player": pydot.Node("Player", shape='ellipse', style='filled', fillcolor='lightblue'),
    "Player_Attributes": pydot.Node("Player_Attributes", shape='ellipse', style='filled', fillcolor='lightblue'),
    "League": pydot.Node("League", shape='ellipse', style='filled', fillcolor='lightblue')
}

# Add nodes to the graph
for node in nodes.values():
    graph.add_node(node)

# Define relationships
edges = [
    ("Team", "Match"),
    ("Match", "Player"),
    ("Player", "Player_Attributes"),
    ("League", "Match")
]

# Add edges to the graph
for edge in edges:
    graph.add_edge(pydot.Edge(nodes[edge[0]], nodes[edge[1]], dir='forward'))

# Create the visualization
png_str = graph.create_png(prog='dot')

# Display the image in Jupyter Notebook
image = Image(png_str)
display(image)
```

```python
# Team Performance Analysis: Assess the performance of each team across different leagues and seasons.
team_performance_query = """
SELECT t.team_long_name as 'Team Name', l.name as 'League Name', m.season as 'Season',
       COUNT(*) as 'Matches Played',
       SUM(CASE WHEN m.home_team_api_id = t.team_api_id AND m.home_team_goal > m.away_team_goal THEN 1
                WHEN m.away_team_api_id = t.team_api_id AND m.away_team_goal > m.home_team_goal THEN 1
                ELSE 0 END) as Wins,
       ROUND(SUM(CASE WHEN m.home_team_api_id = t.team_api_id AND m.home_team_goal > m.away_team_goal THEN 1
                     WHEN m.away_team_api_id = t.team_api_id AND m.away_team_goal > m.home_team_goal THEN 1
                     ELSE 0 END) * 100.0 / COUNT(*), 2) as win_percentage
FROM Match as m
JOIN Team as t ON m.home_team_api_id = t.team_api_id OR m.away_team_api_id = t.team_api_id
JOIN League as l ON m.league_id = l.id
GROUP BY t.team_long_name, l.name, m.season
ORDER BY win_percentage DESC;
"""
team_performance_df = pd.read_sql_query(team_performance_query, connection)
team_performance_df
```

```python
# Top Winners in Leagues: Identify the teams with the most wins in each league for every season.
top_winners_query = """
WITH MatchResults AS (
    SELECT
        l.name AS league_name,
        m.season,
        CASE
            WHEN m.home_team_goal > m.away_team_goal THEN ht.team_long_name
            WHEN m.home_team_goal < m.away_team_goal THEN at.team_long_name
            ELSE 'Draw'
        END as match_winner
    FROM Match as m
    JOIN Team as ht ON m.home_team_api_id = ht.team_api_id
    JOIN Team as at ON m.away_team_api_id = at.team_api_id
    JOIN League as l ON m.league_id = l.id
),
RankedWinners AS (
    SELECT
        league_name,
        season,
        match_winner,
        COUNT(match_winner) OVER (PARTITION BY league_name, season, match_winner) as wins
    FROM MatchResults
    WHERE match_winner != 'Draw'
)
SELECT DISTINCT
    league_name,
    season,
    match_winner,
    wins
FROM RankedWinners
WHERE match_winner IN (
    SELECT match_winner
    FROM RankedWinners rw
    WHERE rw.league_name = RankedWinners.league_name AND rw.season = RankedWinners.season
    ORDER BY wins DESC
    LIMIT 5
)
ORDER BY league_name, season, wins DESC;
"""
top_winners_df = pd.read_sql_query(top_winners_query, connection)
top_winners_df
```

```python
# Top 5 Players by Season: Determine the top 5 players each season based on their overall rating.
top_5_players_query = """
SELECT DISTINCT subquery.player_name, subquery.overall_rating, subquery.season ,subquery.Rank
FROM (
    SELECT DISTINCT p.player_name, f.overall_rating, strftime('%Y', f.date) AS season,
        ROW_NUMBER() OVER(PARTITION BY strftime('%Y', f.date) ORDER BY f.overall_rating DESC) AS Rank
    FROM player AS p
    INNER JOIN player_attributes AS f
    USING(player_api_id)
    GROUP BY player_name ,season
) AS subquery
WHERE subquery.Rank <= 5
GROUP BY subquery.season, subquery.Rank;
"""
top_5_players_df = pd.read_sql_query(top_5_players_query, connection)
top_5_players_df
```

```python
# League Home and Away Wins: Analyze league matches to calculate the total number of home and away wins
league_wins_query = """
SELECT
    l.name as league_name,
    SUM(CASE WHEN m.home_team_goal > m.away_team_goal THEN 1 ELSE 0 END) as home_wins,
    SUM(CASE WHEN m.away_team_goal > m.home_team_goal THEN 1 ELSE 0 END) as away_wins
FROM
    Match as m
JOIN
    League as l ON m.league_id = l.id
GROUP BY
    l.name;
"""
league_wins_df = pd.read_sql_query(league_wins_query, connection)
league_wins_df
```

```python
# Evaluate the distribution of player ages and their average, median and modal ratings in different age groups
import numpy as np
from scipy import stats

# Fetch data
query = """
SELECT
    CASE
        WHEN (2016 - CAST(strftime('%Y', p.birthday) AS INTEGER)) BETWEEN 18 AND 23 THEN '18-23'
        WHEN (2016 - CAST(strftime('%Y', p.birthday) AS INTEGER)) BETWEEN 24 AND 30 THEN '24-30'
        WHEN (2016 - CAST(strftime('%Y', p.birthday) AS INTEGER)) BETWEEN 31 AND 36 THEN '31-36'
        ELSE '36+'
    END as age_group,
    pa.overall_rating
FROM Player p
JOIN Player_Attributes pa ON p.player_api_id = pa.player_api_id;
"""
df = pd.read_sql_query(query, connection)

# Calculate average, median, and mode
age_analysis = df.groupby('age_group')['overall_rating'].agg(['count', 'mean', 'median'])
age_analysis['mode'] = df.groupby('age_group')['overall_rating'].agg(lambda x: stats.mode(x)[0])
age_analysis.rename(columns={'count': 'Number of Players', 'mean': 'Average Ratings', 'median': 'Median Ratings', 'mode': 'Mode Ratings'}, inplace=True)

age_analysis
```
